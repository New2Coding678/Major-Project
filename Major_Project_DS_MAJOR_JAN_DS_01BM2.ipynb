```python
df = df[["location", "date", "new_cases"]]
```

```python
df = df.reset_index()
df["index"] = df.index
```

```python
df
```

```python
df.rename(columns = {'new_cases' : 'y','date':'ds'},inplace = True)
```

```python
df
```

```python
df_new = df[['ds','y']]
df_new
```

```python
df_new.tail(1)
```

```python
df_new.info()
```

```python
df_new['ds'] = pd.to_datetime(df_new['ds'])
df_new['y'] = df_new['y'].astype(int)
df_new.info()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20,5))
plt.scatter(df_new['ds'],df_new['y'])
plt.show()
```

```python
df_new
import numpy as np
df_new['y'] = np.where(df_new['y']>40000, np.nan, df_new['y'])
df_new

# Giving 2750 as threshold
```

```python
df_new.isnull().sum()
```

```python
# Fb Prophet

import fbprophet

op = fbprophet.Prophet(changepoint_prior_scale=1)
op.fit(df_new)
```

```python
forecast = op.make_future_dataframe(periods = 365 ,freq = 'D')
forecast = op.predict(forecast)
```

```python
forecast.columns
```

```python
forecast[['ds','yhat','yhat_lower','yhat_upper']]
```

```python
op.plot(forecast,xlabel = 'Date',ylabel='New Cases');
```
